```python
import pandas as pd
data = pd.read_csv(r'C:\WORK_IAAC\submissions\Storytelling\fiesta_shapefiles\fiestasWiki_national.csv')
data.head()
```

```python
prov_list = []
for index, row in data.iterrows():
    if '(' in row['localidad']:
        prov = row['localidad'].split('(')[1].replace(')', '').strip()
    else:
        prov = row['localidad']
    prov_list.append(prov)
    row['localidad'] = row['localidad'].split('(')[0].strip()
data['prov'] = prov_list

data.head()
```

```python
data['lat'] = data['lat'].astype(str)

data[['Latitude', 'Longitude']] = data['lat'].dropna().apply(lambda x: pd.Series(x.split('/')[0:2]) if '/' in x else pd.Series([None, None]))

data = data.drop(columns=['lat'])
data.head()
```
