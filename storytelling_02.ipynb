```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
url = "https://es.wikipedia.org/wiki/Anexo:Fiestas_de_Inter%C3%A9s_Tur%C3%ADstico_Internacional_(Espa%C3%B1a)#Fiestas_de_inter%C3%A9s_tur%C3%ADstico_internacional"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table", {"class": "wikitable sortable"})

header = [th.text.strip() for th in table.find("tr").find_all("th")]

data = []
for tr in table.find_all("tr")[1:]:
    row = [td.text.strip() for td in tr.find_all("td")]
    data.append(row)

data = pd.DataFrame(data, columns=header)

data.head()
```

```python
data.keys()
```

```python
data = data.drop('Identificador en Wikidata', axis=1)
data = data.drop('Categoría de imágenes en Commons', axis=1)
data = data.drop('Imagen', axis=1)
data = data.rename(columns={"Fiesta": "festival", "Fecha de declaración": "est","Lugar de celebración": "localidad","Fecha de celebración": "date_start","Coordenadas": "lat"})
# Renaming one or more columns using a function that will be applied to each column name. This can be useful if you want to modify all column names in a certain way.
# df = df.rename(columns=lambda x: x.lower().replace(' ', '_'))

data.head()
```

```python
data['lat'] = data['lat'].apply(lambda x: x.split('/')[1].strip())
data['lng'] = data['lat'].apply(lambda x: x.split(',')[1].split('(')[0].strip())
data['est'] = data['est'].apply(lambda x: x.split('[')[0].strip())
data.head()
```

```python

data['lat'] = data['lat'].apply(lambda x: x.split(',')[0].strip())
data.head()
```

```python
#dublicate values to new column from existing on condition
prov_list = []
for index, row in data.iterrows():
    if '(' in row['localidad']:
        prov = row['localidad'].split('(')[1].replace(')', '').strip()
    else:
        prov = row['localidad']
    prov_list.append(prov)
    row['localidad'] = row['localidad'].split('(')[0].strip()
data['prov'] = prov_list

data.head()
```

```python
data.insert(5, 'type', '')
data = data.assign(tags = [""] * len(data))
data = data.assign(comunidad = [""] * len(data))
data = data.assign(presupuesto_2022 = [""] * len(data))
data = data.assign(turistas_2022 = [""] * len(data))

data.head()
```

```python
def add_carnival_dates(date):
    if date == 'Carnaval':
        start = datetime(2023, 2, 16)
        fin = datetime(2023, 2, 22)
        return pd.Series({'start': start, 'fin': fin})
    if date == 'Semana Santa':
        start = datetime(2023, 4, 2)
        fin = datetime(2023, 7, 9)
        return pd.Series({'start': start, 'fin': fin})
    else:
        return pd.Series({'start': None, 'fin': None})

data = pd.concat([data, data['date_start'].apply(add_carnival_dates)], axis=1)
data['start'] = pd.to_datetime(data['start'])
data['fin'] = pd.to_datetime(data['fin'])
 
data.head()
```
