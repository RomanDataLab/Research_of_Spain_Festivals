```python
import requests
from bs4 import BeautifulSoup

url = "https://www.larazon.es/espana/20211224/rgveqijauvcg7hkoyuzskngebe.html"

# Make a request to the website
response = requests.get(url)
fest = []
# Check if the request was successful
if response.status_code == 200:
    # Get the HTML content of the page
    html_content = response.text

    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(html_content, "html.parser")

    # Find all the `li` tags that contain a `span` tag
    li_tags = soup.find_all("li")
    for li_tag in li_tags:
        span_tag = li_tag.find("span")
        if span_tag:
            # Get the text inside the `span` tag
            text = span_tag.text
            fest.append(text)
            # Cut the text until the second `b` tag
            index = text.find("</b>", text.find("</b>") + 1)
            if index != -1:
                text = text[:index]
                fest.append(text)
            
else:
    print("Failed to retrieve the webpage")
print(fest)
```

```python
fest2 = fest[9:]
fest3 = fest2[:-10]
print(fest3)
```

```python
fest4 = [item.split(")", 1)[0] + "." for item in fest3]

print(fest4)
```

```python
for item in fest4:
    parts = item.split("/")
    print(parts[1])
```

```python
import pandas as pd
fest5 = [f[:-1] for f in fest4] # remove '.' from the end of each string

data = []
for f in fest5:
    part1 = f.split("/")
    part2 = part1[1].split("(")
    if "," in part2[1]:
        part3 = part2[1].split(",")
    else:
        part3 = part2[1] + "," + part1[0]
        part3 = part3.split(",")
    province = part1[0].strip().lower().capitalize()
    fest_name = part2[0].strip()
    date_start = part3[0].strip()
    localidad = part3[1].strip().lower().capitalize()
    if "-" in date_start:
        date_start, date_fin = date_start.split("-")
    else:
        date_fin = date_start
    date_start = date_start.replace("de ", "")
    date_fin = date_fin.replace("de ", "")
    localidad = localidad.replace(" y", ".")
    data.append({"province": province, "fest_name": fest_name, "date_start": date_start, "date_fin": date_fin, "localidad": localidad})

df = pd.DataFrame(data, columns=["province", "fest_name", "date_start", "date_fin", "localidad"])
df.head()
```

```python
import geocoder

df['lat'] = None
df['lng'] = None

# Loop through each row of the dataframe and get the latitude and longitude for each address
for index, row in df.iterrows():
    address = row['localidad'][0] + row['province']
    
    g = geocoder.arcgis(address)
    df.at[index, 'lat'] = g.lat
    df.at[index, 'lng'] = g.lng
    print (g.lat,g.lng)
```
